```python
import pandas as pd
import seaborn as sns
```

```python
input = pd.read_csv('../saved-views/year_month-labeled-customers.csv',
                    parse_dates=['year_month'],
                    date_format='%Y%m')
input.head()
```

```python
year_month_list = pd.unique(input['year_month'])

# Will contain the list of unique customers as we go through each month
customer_list = []

# Will track the number of unique customers by year_month
customer_growth = pd.DataFrame(columns=['year_month', 'unique_customers'])

for current_year_month in year_month_list:
    # Pull the current year_month's customer numbers
    current_year_month_customers = \
        input.loc[input['year_month'] == current_year_month, 'customerNumber']

    # Add 
    for customer_number in current_year_month_customers:
        if customer_number not in customer_list:
            customer_list.append(customer_number)
    
    customer_growth.loc[customer_growth.shape[0] + 1, :] = \
        {'year_month': current_year_month,
         'unique_customers': len(customer_list)}
```

```python
sns.set_style('whitegrid')
ax = sns.lineplot(data=customer_growth, x='year_month', y='unique_customers')

# Plot formattting
ax.set_xticks(['2003-01', '2003-07', '2004-01', '2004-07', '2005-01', '2005-07'])
ax.set_xticklabels(['Jan\n2003', 'Jul\n2003', 'Jan\n2004', ' Jul\n2004', 'Jan\n2005', 'Jul\n2005'])
ax.set_xlabel('Date')
ax.set_ylabel('Number of Unique Customers')
ax.set_xlim(pd.to_datetime('2003-01-01'), pd.to_datetime('2005-07-01'))
ax.set_ylim(0, 100)
```
